```python
#Problem3: Find Maximum and Minimum Values
def find_max_min(numbers):  

    max_value=numbers[0]
    min_value=numbers[0]
    for num in numbers:     
        if num>max_value:
            max_value=num
        if num<min_value:
            min_value=num
    
    return max_value,min_value
numbers=[45,22,89,10,66]
maximum,minimum=find_max_min(numbers) 

print(f"List:{numbers}")
print(f"Max:{maximum}")
print(f"Min:{minimum}")
```
