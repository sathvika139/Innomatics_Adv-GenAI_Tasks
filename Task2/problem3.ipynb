```python
#Problem3: Find Maximum and Minimum Values

def find_max_min(numbers):   #it takes a list of integers and returns the maximum and minimum values 

    max_value = numbers[0]   #Initialize max and min with the first element
    min_value = numbers[0]
  
    for num in numbers:      #Iterate through the list to find max and min
        if num > max_value:
            max_value = num
        if num < min_value:
            min_value = num
    
    return max_value, min_value

numbers = [45, 22, 89, 10, 66]

maximum, minimum = find_max_min(numbers)  #function call to find the maxim and min values in list

# Display results
print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```
