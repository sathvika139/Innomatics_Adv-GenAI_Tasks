```python
#Problem1: Unique Words in a Sentence

def find_unique_words(sentence):
 
    words = sentence.split() #Split the sentence into words using whitespace as a delimiter
    unique_words = set(words) #Use a set to store unique words, as sets automatically handle duplicates
    
    return unique_words

sentence = input("Enter sentence: ")

unique_words = find_unique_words(sentence) #function call to find unique words in the input sentence

print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```
