```python
#Problem5: Calculate Attendance Percentage

def calculate_attendance_percentage(attendance): #it calculates the attendance percentage from a list containing 'P'(Present) and 'A'(Absent)
    
    total_days = len(attendance)
    present_days = 0    #Initialize counter for present days
    
    for status in attendance:  #Iterate through the attendance list 
        if status == "P":      #Count the number of present days
            present_days += 1
    
    attendance_percentage = (present_days / total_days) * 100
    
    return attendance_percentage  #

attendance = ["P", "P", "A", "P", "P"]

percentage = calculate_attendance_percentage(attendance)  #Call the function to calculate attendance percentage

print(f"Attendance Percentage: {percentage}")
```
