```python
#Problem5: Calculate Attendance Percentage
def calculate_attendance_percentage(attendance): 
    total_days=len(attendance)
    present_days=0    
    for status in attendance: 
        if status=="P":     
            present_days+=1
    attendance_percentage=(present_days/total_days)*100
    return attendance_percentage  

attendance=["P","P","A","P","P"]
percentage=calculate_attendance_percentage(attendance)  
print(f"Attendance Percentage: {percentage}")
```
