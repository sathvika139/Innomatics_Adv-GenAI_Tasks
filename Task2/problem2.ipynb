```python
#Problem2: Highest Salary from Employee Data

def find_highest_salary(employees):
    """ takes a dictionary of employees and their salaries and returns the employee with the highest salary and the salary amount """

    highest_paid_employee = max(employees, key=employees.get) #Find the key(employee name) with the maximum value(salary)
    
    highest_salary = employees[highest_paid_employee]    #Get the salary of highest paid employee
    
    return highest_paid_employee, highest_salary

employees = {
    "Ravi": 75000,
    "Anita": 68000,  
    "Kiran": 72000
}

name, salary = find_highest_salary(employees)  # Call the function to find the highest salary and the corresponding employee name

print(f"Highest Salary: {name} - {salary}")
```
