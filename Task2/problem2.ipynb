```python
#Problem2: Highest Salary from Employee Data
def find_highest_salary(employees):

    highest_paid_employee=max(employees,key=employees.get) 
    highest_salary=employees[highest_paid_employee]    
    return highest_paid_employee,highest_salary

employees={
    "Ravi": 75000,
    "Anita": 68000,  
    "Kiran": 72000
}
name,salary=find_highest_salary(employees)  
print(f"Highest Salary:{name}-{salary}")
```
