```python
#Problem9: Check if a Key Exists in a Dictionary

def check_key_exists(data, key):  #This function checks whether a given key exists
   
    if key in data:   #Check if the key exists in the dictionary using the 'in' keyword
        return True
    else:
        return False

employees = {
    "Employee": "Ravi",
    "Salary": 75000,
    "Department": "IT"
}

key_to_check = "Employee"

if check_key_exists(employees, key_to_check):   #Call the function to check if the key exists
    print("Employee exists")
else:
    print("Employee does not exist")
```
