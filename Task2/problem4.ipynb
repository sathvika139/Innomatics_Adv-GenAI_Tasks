```python
#Problem4: Count Products Above a Price Threshold

def count_products_above_threshold(prices, threshold=1000): #it counts how many product prices are greater than the given threshold
    count = 0  #Initialize counter
   
    for price in prices:       #Iterate through the list of prices and count how many are above the threshold
        if price > threshold:
            count += 1
    
    return count

prices = [450, 1200, 899, 1500, 300]

result = count_products_above_threshold(prices)  #function call to count products above the threshold of 1000

print(f"Products above 1000: {result}")
```
