```python
#Problem6:Remove Duplicate Phone Numbers

def remove_duplicate_phone_numbers(phone_numbers): #it takes a list of phone numbers and returns a set containing only unique phone numbers.
   
    unique_phone_numbers = set(phone_numbers)  #Convert the list to a set to remove duplicates
    
    return unique_phone_numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = remove_duplicate_phone_numbers(phone_numbers) #Call the function to remove duplicate phone numbers

print(f"Unique phone numbers: {unique_numbers}")
```
