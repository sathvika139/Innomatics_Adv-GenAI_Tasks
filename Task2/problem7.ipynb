```python
#Problem7: Count Character Frequency

def count_character_frequency(text):  # counts how many times each character appears in the given string.
    
    frequency = {}  # Dictionary to store character counts
    
    for char in text:   #Iterate through each character in the input string
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1
    
    return frequency

text = "python"

result = count_character_frequency(text) #function call to count character frequency

print(result)
```
