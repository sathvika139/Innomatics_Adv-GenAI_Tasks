```python
#program to filter premium crop prices based on a threshold value
def premium_crop_filter(prices,threshold=2000):
    premium_crops=[]
    
    for price in prices:
        if price>threshold:
            premium_crops.append(price)
    print("Premium Crops:",premium_crops)

crop_prices=[1800,2500,1500,3200,1900]
premium_crop_filter(crop_prices)
```
