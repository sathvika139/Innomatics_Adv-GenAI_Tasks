```python
#program to check application health based on error count
def application_health_checker(error_count):
    if error_count==0:
        status="Healthy"
    elif error_count<=5:
        status="Minor Issues"
    else:
        status="Critical Issues"
    print("Error Count:",error_count)
    print("System Status:",status)

current_errors=7
application_health_checker(current_errors)
```
