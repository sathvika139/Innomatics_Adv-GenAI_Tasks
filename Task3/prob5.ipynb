```python
# This code analyzes a list of daily sales figures to identify any significant spikes in sales. It calculates the average sales and sets a threshold at 30% above the average. 
# #Then, it iterates through the sales data to find and print any days where the sales exceeded this threshold, indicating a spike in sales.
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.3

print("Sales Spikes Detected:")
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```
