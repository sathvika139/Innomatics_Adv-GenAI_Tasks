```python
# This code counts the frequency of each word in a given query and identifies the words that are repeated 
query = "Buy mobile phone buy phone online"
query = query.lower()

for char in ",.!?;:":
    query = query.replace(char, "")
words = query.split()
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

repeated_words = {word: count for word, count in word_count.items() if count > 1}
print(repeated_words)
```
