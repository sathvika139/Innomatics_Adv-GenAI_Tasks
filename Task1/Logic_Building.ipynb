```python
#Pass/fail Analyser
marks = [45, 78, 90, 33, 60]
p=0
f=0
for i in marks:
  if i>50:
    p=p+1
  else :
    f=f+1

print("passed students:",p)
print("Failed students:",f)
```

```python
#simple data cleaner
names=[" Alice ", "bob", " CHARLIE "]
cleaned=[]
for n in names:
    cleaned.append(n.strip().lower())
print("Cleaned Names:",cleaned)
```

```python
#error message detection
logs=["INFO", "ERROR", "WARNING", "ERROR"]
errors= 0
for log in logs:
    if log=="ERROR":
        errors+=1
print("Total ERRORS:",errors)
```
